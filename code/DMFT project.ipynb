```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as spo
```

```python
def f(x,a,b):
    
    return a*x+b
```

```python
def approxderiv(X,Y):  #fit on 3 points to get the approx derivative
    popt,pcov=spo.curve_fit(f,X,Y)
    return popt[0]
```

```python
#J=0 , 1 mean D1=1 , 2 mean D2=2
Sigma1_0=[]
Sigma2_0=[]
U=[0,1,2,3,4,5,6,7]
Matsu=[]
for j in range(len(U)):
    List1=[]
    List2=[]
    Ma=[]
    filename=r"C:\Users\pttai\Desktop\M2\Electronic structure theory\projet DMFT\Data\J0\U"+str(j)+"\Sigma"+str(j)+".dat"
    file = open(filename,'r')
    for L in file.readlines():
        L=L.split('\t')
        Ma.append(float(L[0]))
        List1.append(float(L[2]))
        List2.append(float(L[8]))
    Matsu.append(Ma)    
    Sigma1_0.append(List1)
    Sigma2_0.append(List2)
    file.close()
```

```python
plt.figure(figsize=(12,8))
plt.plot(Matsu[0],Sigma1_0[-5],label='D1=1',marker='o')
plt.plot(Matsu[0],Sigma2_0[-5],label='D2=2',marker='o')
plt.legend()
plt.show()
```

```python
Z1_0=[]
Z2_0=[]
for j in range(len(U)):
    a1=(approxderiv(Matsu[0][:3],Sigma1_0[j][:3]))
    a2=(approxderiv(Matsu[0][:3],Sigma2_0[j][:3]))
    Z1_0.append(1/(1-a1))
    Z2_0.append(1/(1-a2))
for j in range(len(U)):
    Z1_0[j]=Z1_0[j]/max(Z1_0)
    Z2_0[j]=Z2_0[j]/max(Z2_0)
    
    
print(Z1_0[-1])
print(Z2_0[-1])
```

```python
fig=plt.figure(figsize=(14,8))
ax=fig.gca()
plt.plot(U,Z1_0,label='D1=1',marker='o')
plt.plot(U,Z2_0,label='D2=2',marker='o')
#plt.plot(U,[Z2_0[i]-Z1_0[i] for i in range(len(U))])
plt.legend()
plt.grid()
ax.set_xticks(np.arange(0, 7, 0.2))

plt.xlabel('U')
plt.ylabel('Z')
plt.show()
```
